```python
# 수집된 기사에서 가져온 contents를 리스트에 저장
from trafilatura import fetch_url, extract
import json
import requests
from bs4 import BeautifulSoup
from trafilatura import fetch_url, extract

cur_page = 1
def get_contents(cur_page):
    save_contents = []
    
    while(True):
        try:
            url = f'https://s.search.naver.com/p/newssearch/search.naver?cluster_rank=28&de=2024.11.22&ds=2020.11.22&eid=&field=0&force_original=&is_dts=0&is_sug_officeid=0&mynews=0&news_office_checked=&nlu_query=&nqx_theme=%7B%22theme%22%3A%7B%22main%22%3A%7B%22name%22%3A%22movie_info%22%2C%22source%22%3A%22LLM%22%7D%7D%7D&nso=%26nso%3Dso%3Ar%2Cp%3Afrom20201122to20241122%2Ca%3Aall&nx_and_query=&nx_search_hlquery=&nx_search_query=&nx_sub_query=&office_category=0&office_section_code=0&office_type=0&pd=3&photo=0&query=%EB%A7%88%EB%B8%94&query_original=&service_area=0&sort=0&spq=0&start={cur_page}&where=news_tab_api&nso=so:r,p:from20201122to20241122,a:all&'

            result = requests.get(url)
            try:
                result = result.json()
            except json.JSONDecodeError as e:
                print(f'json 디코딩 에러')
                break
            contents = result["contents"]
            content_url = result["nextUrl"]
            
            if not contents:
                break
            if not content_url:
                break

            for content in contents:
                contentt = get_content(content)
                if contentt:
                    save_contents.append(contentt)
            
            cur_page += 10
            print(cur_page)
        except Exception as e:
            print(f"에러 발생")
            break
        
    return save_contents

def get_content(content):
    content_soup = BeautifulSoup(content, 'html.parser')
    news_title = content_soup.find("a", {"class": "news_tit"})
    if news_title and news_title['title']:
        return news_title['title']
    return None

save_contents = get_contents(cur_page)
print(len(save_contents))
```

```python
import pandas as pd

print("총 개수 :", len(save_contents))
df = pd.DataFrame(save_contents, columns=['content'])
df.to_csv('./crawling_datas/news_onw')
```

```python
from kiwipiepy import Kiwi
from collections import Counter

input_wc = ' '.join(df['content'].astype(str))

kiwi = Kiwi()
kiwi.analyze(input_wc)
```

```python
n = []
result = kiwi.analyze(input_wc)
for token in result[0][0]:
    if token.tag in ['NNG', 'NNP']:
        n.append(token.form)
        
n_count = Counter(n)
```
