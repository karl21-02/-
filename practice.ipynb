```python
# 수집된 기사에서 가져온 contents를 리스트에 저장
from trafilatura import fetch_url, extract
import json
import requests
from bs4 import BeautifulSoup
from trafilatura import fetch_url, extract

cur_page = 1
def get_contents(cur_page):
    save_contents = []
    
    while(True):
        url = f'https://s.search.naver.com/p/newssearch/search.naver?cluster_rank=14&de=2024.11.21&ds=2022.11.21&eid=&field=0&force_original=&is_dts=0&is_sug_officeid=0&mynews=0&news_office_checked=&nlu_query=&nqx_theme=%7B%22theme%22%3A%7B%22main%22%3A%7B%22name%22%3A%22movie%22%2C%22score%22%3A%220.845629%22%7D%7D%7D&nso=%26nso%3Dso%3Ar%2Cp%3Afrom20111121to20241121%2Ca%3Aall&nx_and_query=&nx_search_hlquery=&nx_search_query=&nx_sub_query=&office_category=0&office_section_code=0&office_type=0&pd=3&photo=0&query=%EB%84%B7%ED%94%8C%EB%A6%AD%EC%8A%A4+%EB%A7%88%EB%B8%94&query_original=&service_area=0&sort=0&spq=0&start={cur_page}&where=news_tab_api&nso=so:r,p:from20111121to20241121,a:all&'
        result = requests.get(url)
        result = result.json()
        contents = result["contents"]
        content_url = result["nextUrl"]
        
        if not contents:
            break
        if not content_url:
            break

        for content in contents:
            contentt = get_content(content)
            if contentt:
                save_contents.append(contentt)
        
        cur_page += 10
        print(cur_page)
    return save_contents

def get_content(content):
    content_soup = BeautifulSoup(content, 'html.parser')
    news = content_soup.find("a", {"class": "news_tit"})
    if news:
        return news['title']
    return None

save_contents = get_contents(cur_page)
print(len(save_contents))
```
